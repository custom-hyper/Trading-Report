```python
import numpy as np

def daily_sharpe_ratio(df: pd.DataFrame) -> pd.Series:
  day_wise_returns = df.groupby(df.day).pct

  return (np.sqrt(252) * day_wise_returns.mean() / day_wise_returns.std())
```
